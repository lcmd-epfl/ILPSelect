```python
import matplotlib.pyplot as plt
import matplotlib
```

```python
matplotlib.rcParams.update({'font.size': 14})
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2)
reps_local = ["CM", "aCM", "SLATM", "FCHL", "SOAP"]
ratio_local = [0.33333333, 0, 0.1875, 0.23, 0.25]

reps_global = ["aCM", "SLATM", "FCHL", "SOAP"]
ratio_global = [0.27, 0.27, 0.33, 0.27]

axs[0].set_title("Local")
axs[0].bar(reps_local, ratio_local)
axs[0].set_ylabel("Ratio of disconnected solutions")
axs[0].tick_params(axis='x', labelrotation=45)

axs[1].set_title("Global")
axs[1].bar(reps_global, ratio_global, color='orange')
axs[1].tick_params(axis='x', labelrotation=45)

plt.tight_layout()

plt.savefig("../figures/ratios_disconnected.png", dpi=300)
plt.show()
```
